__[문제 1: 첫 번째 사냥]__

마법사는 몬스터를 사냥하기 위해 마법을 사용합니다.

마법을 사용하려면 지능이 50 이상이어야 합니다.

마법사의 지능이 50 이상인지 확인하세요.

```python
wizard_int = 80
if wizard_int >= 50:
    print("마법을 사용할수있습니다.")
```

__[문제 2: 두 번째 사냥]__

전사는 강력한 공격을 하기 위해 자신의 힘을 사용합니다.

```python
worrior_str = 80
worrior_hp = 100
if worrior_str >= 60 and worrior_hp >= 80:
    print("강력한 공격이 가능합니다")
```

__[문제 3: 세 번째 사냥]__

```python
assassin_dex = 80
assassin_str = 40
if assassin_dex >= 70 or assassin_str >= 50:
    print("회피하고 공격할수있습니다.")
```

__[문제 4: 네 번째 사냥]__

힐러는 파티의 체력을 회복시키기 위해 마법을 사용합니다.

```python
cleric_int = 70
cleric_hp = 70
if cleric_int >= 60 and cleric_hp >= 50:
    print("체력회복마법 가능.")
```

__[문제 5: 레벨업]__

마법사: 120

```python
wizard_exp = 120
worrior_exp = 150
assassin_exp =110
cleric_exp = 90
if wizard_exp >= 100:
    print("마법사 레벨업!")
if worrior_exp >= 100:
    print("전사 레벨업!")
if assassin_exp >= 100:
    print("도적 레벨업!")
if cleric_exp >= 100:
    print("힐러 레벨업!")
```
